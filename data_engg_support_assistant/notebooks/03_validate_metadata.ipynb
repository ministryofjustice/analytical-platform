```python
"""
Script Name: validate_metadata_files.py

Description:
    This script searches for all metadata files matching the pattern "*.txt.metadata.json" within the
    ../data/text_chunks directory and its subdirectories. For each metadata file found, it runs a validation
    function to check for issues in the metadata structure or content. Any detected issues are printed to the console
    along with the corresponding file path.

Usage:
    - Ensure the helpers.metada_validater module is available and contains the validate_metadata_file function.
    - Run the script from a directory where the project root is one level above the current working directory.
    - Example: python validate_metadata_files.py

Notes:
    - The script adds the project root to the Python path to allow importing helper modules.
    - Issues are printed in a human-readable format for easy debugging.

Example Output:
    The following examples reflect the exact error strings produced by the validator:

        ../data/text_chunks/guides/intro.txt.metadata.json
          - File size 15234 bytes exceeds 10KB limit
          - Missing 'metadataAttributes' key

        ../data/text_chunks/howto/setup.txt.metadata.json
          - Invalid JSON: Expecting property name enclosed in double quotes: line 12 column 5 (char 231)

        ../data/text_chunks/specs/details.txt.metadata.json
          - 'metadataAttributes' is not a dict

        ../data/text_chunks/cases/example.txt.metadata.json
          - key 'very_long_key_name_...' exceeds 255 characters
          - key'missing_title' has null value
          - key'description' has empty string
          - key'links' has unsupported type list/dict
          - key 'rating' has unsupported type NoneType
          - Value for key 'summary' is too long (1345 chars)

    If no issues are found for a file, it won't be printed.
"""
```

```python
import os
import sys
import glob

# Add project root to Python path so imports from parent directory work
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

# Define the file pattern for metadata files (recursive search in ../data/text_chunks)
metadata_filepath = "../data/text_chunks/**/*.txt.metadata.json"

# Use glob to find all matching metadata files recursively
metadata_files = glob.glob(metadata_filepath, recursive=True)

# Import the metadata validation function from helpers module
from helpers.apug.metada_validator_03 import validate_metadata_file
```

```python
# Iterate over each metadata file found
for file in metadata_files:
    # Validate the metadata file and collect any issues
    issues = validate_metadata_file(file)

    # If issues are found, print the file name and each issue
    if issues:
        print(f"\n {file}")
        for issue in issues:
            print("  -", issue)
```
